### Importando as bibliotecas

```python
import matplotlib.pyplot as plt # Permite a geração de graficos

import numpy as np # Permite trabalhar com matrizes e vetores
```

### Primeiro gráfico com matplotlib

```python
# Criando os vetores(Conjunto de dados)

x = np.array([1,2,3,4,5])

y = np.array([2,4,6,8,10])

# Plotar o gráfico

plt.plot(x,y)
```

### Modificando o tipo dos marcadores

```python
plt.plot(x,y,marker='*')
```

#### Alterando as cores da linha.

```python
plt.plot(x,y,marker ='*', c='g')
```

### Mudando como os pontos são ligados e o tipo de linha.

```python
plt.plot (
    x,y,          # Os dados do gráfico
    marker = '*',  # Tipo de Marcador
    c = 'red',   # Cor da linha
    ds = 'steps',  # Forma de conexão dos pontos
    ls = '--',     # Estilo da Linha
    mec = 'blue',  # Cor dos Marcadores
)
```

### Matplotlib- Adicionando outros elementos

### Podemos ainda adicionar labels aos eixos, título ao gráfico, modificar o tamanho da fonte, dentre outraspossibilidades. Tome cuidado para que seu gráfico não tenha muitos elementos e acabe tirando aatenção do que realmente importa.

```python
plt.plot(x,y)
plt.xlabel('Valores de X', fontsize = 12)
plt.ylabel('Valores de Y', fontsize = 12)

# Adiciona titulo ao Gráfico
plt.title ('Título do gráfico', fontsize = 18, c='purple')

# Altera as cores dos valores dos eixos
plt.xticks(fontsize = 8, c = 'blue')
plt.yticks(fontsize = 8, c = 'orange')
```
