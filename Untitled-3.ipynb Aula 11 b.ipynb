### Exercício 01 - Escreva um programa Python para desenhar uma linha com rótuloadequado no eixo x, eixo y e um título.

Amostra de dados:
x = range(1,50)
y = [value * 3 for value in x]

```python
# Importar a biblioteca

import matplotlib.pyplot as plt

# Gerar os dados para inserir no gráfico

x = range(1,50)
y = [value * 3 for value in x]

print('Valores de X')
print(range(1,50))
print('Valores de Y(Três vezes)')
print(y)
```

```python
# Plotar um gráfico de linha 

plt.plot(x,y)

#Adicionar um titulo ao grafico

plt.title('Draw a line', fontsize = 15, c= 'blue')

# Adicionar titulo aos eixos 

plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
```

### Multiplas curvas no mecanismo gráfico

```python
import matplotlib.pyplot as plt
import numpy as np

# Retorna (Linspace) valores igualmente espaçados em um intervalo
x = np.linspace (0,2*np.pi,100) 
y = np.sin(x) 
z = np.cos(x) 
plt.plot (x,y) 
plt.plot (x,z)
```

### Matplotlib - Múltiplos subgraficos

#

```python
# O comando plt.subplot permite criar grupo de gráfico
# plt. subplot (linhas, colunas, Posição elemento)

plt.subplot(2,1,1) # Primeiro Gráfico
plt.plot(x,y) # Conjunto de dados do primeiro grafico
plt.title('subplots')

plt.subplot(2,1,2) # Segundo gráfico
plt.plot(x,z) # Conjunto de dados do segundo gráfico
```

```python
### Gráficos na Vertical

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('subplots')
plt.subplot(1,2,2)
plt.plot(x,z)
```
